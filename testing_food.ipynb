```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
cnn = tf.keras.models.load_model("/content/drive/MyDrive/food/trained_model.h5")
```

```python
import cv2
image_path  = "/content/drive/MyDrive/food/test/apple/Image_1.jpg"
img = cv2.imread(image_path)
plt.imshow(img)
```

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])
prediction = cnn.predict(input_arr)

```

```python
print(prediction)
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/food/test',
    labels = 'inferred',
    label_mode = 'categorical',
    class_names =None,
    color_mode = 'rgb',
    batch_size = 32,
    image_size = (64,64),
    shuffle = True,
    seed = None,
    validation_split = None,
    subset = None,
    interpolation = 'bilinear',
    follow_links = False,
    crop_to_aspect_ratio = False

)
```

```python
result_index = np.where(prediction[0] == max(prediction[0]))
print(result_index)
```

```python
plt.imshow(img)
```

```python
print("It is a {}".format(test_set.class_names[result_index[0][0]]))
```
